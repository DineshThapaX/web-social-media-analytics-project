# **1. Data Preprocessing**

**Loading Dataset into Dataframe**

```python
import pandas as pd

# Load the dataset
from google.colab import drive
drive.mount('/content/drive/')

# Specify the path to your dataset file
file_path = '/content/drive/MyDrive/SocialMediaAnalyticsProject/covid19_tweets.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)
df
```

**Filter tweets having #brexit hashtag**

```python
# Initially filter for tweets containing 'Brexit' in hashtags
df = df[df['hashtags'].str.contains('Brexit', case=False, na=False)]
df
```

**Count the number of rows in the dataframe**

```python
# Count the number of rows in the filtered DataFrame
num_rows = df.shape[0]  # Alternatively, you can use len(df)

# Print the number of rows
print("Number of tweets containing 'Brexit':", num_rows)
```

**Selecting only necessary columns for our analysis**

```python
df = df[['user_location', 'date', 'text', 'source']]
df
```

**Handling Missing Data**

```python
# For the scope of sentiment analysis, ensure there are no missing values in 'text'
df.dropna(subset=['text'], inplace=True)
df
```

**Text Preprocessing Operations**

```python
import re
# Advanced Text Preprocessing
def clean_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove user @ references and '#' from tweet
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Optionally, remove numbers
    text = re.sub(r'\d+', '', text)
    return text

# Apply text cleaning function
df['cleaned_text'] = df['text'].apply(clean_text)

# Display the cleaned DataFrame
print(df.head())
```

**Remove Raw Text Data as new Text Data is in cleaned_text column**

```python
if 'text' in df.columns:
    df.drop('text', axis=1, inplace=True)  # Remove the 'text' column from the DataFrame
    print("Column 'text' removed successfully.")
else:
    print("Column 'text' not found in the DataFrame.")

# Display the first few rows of the DataFrame to confirm
print(df.head())
```

**Save the cleaned data from dataframe to CSV file**

```python
# Save the cleaned DataFrame back to CSV if needed
df.to_csv('/content/drive/MyDrive/SocialMediaAnalyticsProject/cleaned_covid19_tweets.csv', index=False)
```

# **2. Sentiment Analysis**

**Loading Cleaned/Preprocessed Data for Analysis**

```python
# Specify the path to your dataset file
claned_file_path = '/content/drive/MyDrive/SocialMediaAnalyticsProject/cleaned_covid19_tweets.csv'

# Load the dataset into a DataFrame
cleaned_df = pd.read_csv(claned_file_path)
cleaned_df
```

**Sentiment Analysis**

```python
# Using TextBlob Library
from textblob import TextBlob
```

```python
# Define a function to get sentiment polarity and subjectivity
def get_sentiment(text):
    if pd.isna(text):
        return None, None  # Return None if text is NaN
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# Apply the sentiment analysis on the 'cleaned_text' column
cleaned_df['polarity'], cleaned_df['subjectivity'] = zip(*cleaned_df['cleaned_text'].apply(get_sentiment))

# Categorize sentiments into positive, neutral, and negative based on polarity
def categorize_sentiment(polarity):
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply categorization
cleaned_df['sentiment_category'] = cleaned_df['polarity'].apply(categorize_sentiment)

# Display the first few rows to verify
print(cleaned_df[['cleaned_text', 'polarity', 'subjectivity', 'sentiment_category']].head())
```

**Sentiment Distribution Visualization in Bar Diagram**

```python
import matplotlib.pyplot as plt

# Visualization: Plot the distribution of sentiment categories
# Count the occurrences of each sentiment category
sentiment_counts = cleaned_df['sentiment_category'].value_counts()

colors = ['#57e32c', '#5a8def', '#f24444']  # green for positive, blue for neutral, red for negative

# Create a bar chart
plt.figure(figsize=(8, 4))
sentiment_counts.plot(kind='bar', color=colors)
plt.title('Distribution of Sentiment Categories')
plt.xlabel('Sentiment Category')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Keep the category labels horizontal
plt.show()
```

**Sentiment Distribution Visualization in Pie Chart**

```python
# Define colors for the pie chart
colors = ['#57e32c', '#5a8def', '#f24444']  # green for positive, blue for neutral, red for negative

# Create a pie chart
plt.figure(figsize=(8, 8))  # Set the size of the figure
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Distribution of Sentiment Categories')
plt.show()
```

**Frequency Distribution of Words in Sentiments**

```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk import FreqDist, word_tokenize
from nltk.corpus import stopwords
import nltk

# Ensure necessary nltk resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Function to tokenize and remove stopwords
def process_text(text):
    tokens = word_tokenize(text.lower())  # Tokenize and convert to lower case
    words = [word for word in tokens if word.isalpha()]  # Remove punctuation
    words = [word for word in words if word not in stopwords.words('english')]  # Remove stopwords
    return words

# Apply text processing to each tweet and append results to new column
cleaned_df['words'] = cleaned_df['cleaned_text'].apply(process_text)

# Prepare data for each sentiment category
positive_words = [word for sublist in cleaned_df[cleaned_df['sentiment_category'] == 'positive']['words'] for word in sublist]
negative_words = [word for sublist in cleaned_df[cleaned_df['sentiment_category'] == 'negative']['words'] for word in sublist]
neutral_words = [word for sublist in cleaned_df[cleaned_df['sentiment_category'] == 'neutral']['words'] for word in sublist]

# Generate frequency distributions
positive_freq = FreqDist(positive_words)
negative_freq = FreqDist(negative_words)
neutral_freq = FreqDist(neutral_words)

# Function to plot word frequencies
def plot_word_freq(freq_dist, title):
    words = [word[0] for word in freq_dist.most_common(20)]
    counts = [word[1] for word in freq_dist.most_common(20)]
    plt.figure(figsize=(10, 5))
    plt.bar(words, counts)
    plt.title(title)
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.show()

# Plotting word frequency for each sentiment
plot_word_freq(positive_freq, 'Top 20 Words in Positive Tweets')
plot_word_freq(negative_freq, 'Top 20 Words in Negative Tweets')
plot_word_freq(neutral_freq, 'Top 20 Words in Neutral Tweets')
```
